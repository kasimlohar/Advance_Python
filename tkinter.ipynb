```python
# Application: Guess the Number Game
import tkinter as tk
import random

class GuessTheNumberApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Guess the Number")
        self.target_number = random.randint(1, 100)
        
        # Create GUI elements
        self.label = tk.Label(root, text="Guess a number between 1 and 100", font=("Arial", 14))
        self.label.pack(pady=10)
        
        self.entry = tk.Entry(root, font=("Arial", 14))
        self.entry.pack(pady=10)
        
        self.submit_button = tk.Button(root, text="Submit", font=("Arial", 14), command=self.check_guess)
        self.submit_button.pack(pady=5)
        
        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=10)
        
        self.restart_button = tk.Button(root, text="Restart", font=("Arial", 14), command=self.restart_game)
        self.restart_button.pack(pady=5)
        
    def check_guess(self):
        try:
            guess = int(self.entry.get())
            if guess < self.target_number:
                self.result_label.config(text="Too low! Try again.", fg="blue")
            elif guess > self.target_number:
                self.result_label.config(text="Too high! Try again.", fg="blue")
            else:
                self.result_label.config(text="Correct! You guessed it!", fg="green")
        except ValueError:
            self.result_label.config(text="Please enter a valid number.", fg="red")
    
    def restart_game(self):
        self.target_number = random.randint(1, 100)
        self.entry.delete(0, tk.END)
        self.result_label.config(text="")
        self.label.config(text="Guess a number between 1 and 100")

# Run the app
if __name__ == "__main__":
    root = tk.Tk()
    app = GuessTheNumberApp(root)
    root.mainloop()
```

```python
# APPLICATION: Sentiment Analysis App
import tkinter as tk
from textblob import TextBlob

class SentimentAnalysisApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Sentiment Analysis App")
        self.root.geometry("400x300")
        
        # Create GUI elements
        self.label = tk.Label(root, text="Enter text to analyze sentiment:", font=("Arial", 12))
        self.label.pack(pady=10)
        
        self.text_input = tk.Text(root, height=6, width=40, font=("Arial", 12))
        self.text_input.pack(pady=10)
        
        self.analyze_button = tk.Button(root, text="Analyze Sentiment", font=("Arial", 12), command=self.analyze_sentiment)
        self.analyze_button.pack(pady=5)
        
        self.result_label = tk.Label(root, text="", font=("Arial", 12))
        self.result_label.pack(pady=10)
        
    def analyze_sentiment(self):
        # Get user input
        user_text = self.text_input.get("1.0", tk.END).strip()
        
        if user_text:
            # Perform sentiment analysis
            blob = TextBlob(user_text)
            polarity = blob.sentiment.polarity
            
            # Determine sentiment type
            if polarity > 0:
                sentiment = "Positive"
                color = "green"
            elif polarity < 0:
                sentiment = "Negative"
                color = "red"
            else:
                sentiment = "Neutral"
                color = "blue"
            
            # Display result
            self.result_label.config(text=f"Sentiment: {sentiment} (Polarity: {polarity:.2f})", fg=color)
        else:
            self.result_label.config(text="Please enter some text.", fg="orange")

# Run the app
if __name__ == "__main__":
    root = tk.Tk()
    app = SentimentAnalysisApp(root)
    root.mainloop()
```

```python
# APPLICATION: Addition of Two Numbers
import tkinter as tk
from tkinter import messagebox

def calculate_sum():
    try:
        # Retrieve numbers from the entry fields
        num1 = float(num1_entry.get())
        num2 = float(num2_entry.get())
        
        # Calculate the sum
        result = num1 + num2
        
        # Update the result label
        result_label.config(text=f"Result: {result}", fg="green")
    except ValueError:
        # Handle invalid inputs
        messagebox.showerror("Invalid Input", "Please enter valid numbers.")
        result_label.config(text="")

# Create the main application window
root = tk.Tk()
root.title("Addition of Two Numbers")
root.geometry("300x200")

# Label and input for the first number
label1 = tk.Label(root, text="Enter first number:", font=("Arial", 12))
label1.pack(pady=5)
num1_entry = tk.Entry(root, font=("Arial", 12))
num1_entry.pack(pady=5)

# Label and input for the second number
label2 = tk.Label(root, text="Enter second number:", font=("Arial", 12))
label2.pack(pady=5)
num2_entry = tk.Entry(root, font=("Arial", 12))
num2_entry.pack(pady=5)

# Button to calculate the sum
add_button = tk.Button(root, text="Add", font=("Arial", 12), command=calculate_sum)
add_button.pack(pady=10)

# Label to display the result
result_label = tk.Label(root, text="Result will be displayed here", font=("Arial", 12))
result_label.pack(pady=10)

# Run the application
root.mainloop()
```
